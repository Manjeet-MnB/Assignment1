```python
#import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
```

```python
#get the data
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
y=df['species']
```

```python
x.head()
```

```python
y.head()
```

```python
y.value_counts()
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
#change the dataset in to 3 dimensions
from sklearn.decomposition import PCA
pca=PCA(n_components=3)
x=pca.fit_transform(x)
```

```python
#visualize the data
figure=plt.figure(figsize=(10,8))
plt.clf()
ax=Axes3D(figure,rect=[0,0,1,1],elev=50,azim=136)
plt.cla()
ax.scatter(x[:,0],x[:,1],x[:,2],c=y,edgecolor='k',cmap='RdYlGn')
plt.show()
```
